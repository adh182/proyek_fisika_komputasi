# G Harmonis Sederhana 1D
Perbandingan hasil exact dan numerik (Forward Finite Difference)

```python
from numpy import zeros, linspace, pi, cos, array
import matplotlib.pyplot as plt
```

```python
omega = 2 #sqrt(k/m)
P = 2*pi/omega #priode getaran
dt = P/2000
#semakin kecil dt maka hasil semakin mendekati hasil exact
T = 5*P #waktu maksimal dgn jumlah 5 gelombang
N_t = int(round(T/dt)) #membulatkan jumlah N
t  = linspace(0, N_t*dt, N_t+1) #mencacah waktu dari 0 hingga n*dt sebanyak n+1

u = zeros(N_t+1)
v = zeros(N_t+1)

#inisialisasi
X_0 = 2
u[0] = X_0
v[0] = 0
```

```python
#penyelesaian fungsi PD dengan finite difference
for n in range(N_t):
    u[n+1] = u[n] + dt*v[n]
    v[n+1] = v[n] - dt*(omega**2)*(u[n])
```

```python
#perintah ploting
fig = plt.figure()
l1, l2 = plt.plot(t, u, 'b-', t, X_0*cos(omega*t), 'r--')
fig.legend((l1, l2), ('numerical', 'exact'), 'upper left')
plt.xlabel('waktu')
plt.ylabel('posisi')
plt.show()
```
